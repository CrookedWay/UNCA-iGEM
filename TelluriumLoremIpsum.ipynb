C1 describes the problem space.

The J's define the relationships between them.

```python
model = te.loada('''
    compartment C1;
    C1 = 0.5;
    species S1, S2, S3, S4;

    S1 = 10000.0;
    S2 = 0.0;
    S3 = 0.0;
    S4 = 0.0;
    
    S1 in C1; S2 in C1; S3 in C1; S4 in C1;
    
    J1: S1 -> S2; k1*S1;
    J2: S2 -> S3; k2*S2;
    J3: S3 -> S4; k3*S3;
    J4: S4 -> S3; k4*S4;

    k1 = 0.6;
    k2 = 0.2;
    k3 = 0.5;
    k4 = 0.1;


''')
```
